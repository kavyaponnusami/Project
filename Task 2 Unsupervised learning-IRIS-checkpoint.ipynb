# Dimentionality reduction

```python
# Lda

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA


lda = LDA(n_components=2)
df['lda1'] = lda.fit(X, y).transform(X)[:, 0]
df['lda2'] = lda.fit(X, y).transform(X)[:, 1]
```

```python
setosa = df.query("Species=='Iris-setosa'")
versicolor = df.query("Species=='Iris-versicolor'")
virginica = df.query("Species=='Iris-virginica'")

# ax.set_aspect('equal')


plt.figure(2)
ax1 = sns.kdeplot(setosa.lda1, setosa.lda2, cmap="Greens",
          shade=True, shade_lowest=False)
ax1 = sns.kdeplot(versicolor.lda1, versicolor.lda2, cmap="Reds",
          shade=True, shade_lowest=False)
ax1 = sns.kdeplot(virginica.lda1, virginica.lda2, cmap="Blues",
          shade=True, shade_lowest=False)

red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
green = sns.color_palette("Greens")[-2]
ax1.text(7, 1.0, "virginica", size=16, color=blue)
ax1.text(-1, 1.5, "versicolor", size=16, color=red)
ax1.text(-5.8, 0.5, "setosa", size=16, color=green)
ax1.set_title('LDA')
```

# Clustering

```python
x=df.iloc[:,:-1].values
from sklearn.cluster import KMeans
wcss=[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

K is 3 since the curve is bend at 3.

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'pink', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'purple', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'yellow', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'red', label = 'Centroids')

plt.legend()
```
